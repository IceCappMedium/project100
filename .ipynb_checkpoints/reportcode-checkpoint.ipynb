# Producing a classifier to determine whether a patient has Heart Disease

## Introduction:
Heart disease, also known as coronary artery disease, is a common cardiovascular ailment that affects the main blood arteries that supply the heart muscle. Plaques, or accumulations of cholesterol within the heart arteries, are usually the cause of this condition. This accumulation, known as atherosclerosis, reduces blood flow to the heart and other parts of the body, which may cause a heart attack, angina (chest discomfort), or stroke. Elevated blood pressure, elevated low-density lipoprotein (LDL) cholesterol, diabetes, tobacco use, obesity, a poor diet, and insufficient physical exercise are some of the main risk factors for heart disease and stroke.

With this project, we wanted to determine the accuracy of the classification of the heart disease prognosis given specific predictors from the given data set. However, note that we additionally also figured out the recall and precision for a better understanding of our work and results. In particular we looked to find out if an undiagnosed patient was likely to have heart disease given age, cholesterol, resting blood pressure, maximum heart rate, and ST depression induced by exercise relative to rest (where ‘ST’ relates to positions on the ECG plot)

We chose to work with the Hungary data set from the Heart Disease, and given the variables mentioned above, make classifications on undiagnosed patients. The reason for this was that the other datasets within heart disease had either lots of missing data or not enough viable information in general to be a good source for classification, as to why we chose the Hungary data set from Heart Disease.

```r
library(tidyverse)
library(tidymodels)
library(repr)
library(dplyr)
library(purrr)
library(gridExtra)

#install.packages("htmltools") in order to updated to newer version needed for ggmosaic
library(ggmosaic, lib.loc = "packages")
```

```r
#download the heart and disease data from online 

hd_data <- read_delim("https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.hungarian.data",delim=",", 
                      col_names = c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", 
                                    "ca", "thal", "num")) #assign the column name accordingly with the given column names from online data base
cat("Table 1: Unfiltered Heart Disease Hungarian Data")
head(hd_data)

                  
```

What we have done in the table above is simply bring in the heart disease data for Hungary into jupyter. We've esentially just downloaded the data frame from online. We can see that there are a lot of missing data within the code where there are question marks. 

```r
#Choosing the predicting variables that we are going to use to predict the class type(avoid using slop,ca,thl because they have missing values)
set.seed(1234)
process_hd_data<-hd_data|>
#choosing the predicting variables and class variable to stay within our data frame
select(c(age,chol,trestbps,thalach,oldpeak,num))|>
#filter out any missing values that is labelled with "?"
filter(age!="?",chol!="?",trestbps!="?",thalach!="?",oldpeak!="?",num!="?")|>
#change the type of the data to dbl because the predicting variables we are predicting should be integer instead of character
mutate(across(c(age, chol, trestbps, thalach, oldpeak, num), as.integer))|>
#add a new column containing the Diagnosis of the patient as either healthy or unhealthy
mutate(Diagnosis= case_when(num == 0 ~ "Healthy",num == 1 ~ "Unhealthy"))|>
#change the type of the class variable to factor because the class variable we are predicting should be categorical
mutate(Diagnosis=as.factor(Diagnosis))

cat("Table 2: Filtered Heart Disease Hungarian Data")
head(process_hd_data)
```

In order to resolve the issue of missing data from the previous cell above, we performed some filtering, and also added a new coloumn called 'Diagnosis' in order to have the table be more understandable than just using the numbers of 0, and 1. 0 means healthy, and the patient doesn't have heart diease, while 1 means unhealthy, and the patient does have heart disease. We've specifically filtered for columns containing our predicting and response/class variables, as other information is not necessary. We've also turned our newly made Diagnosis column into a factor, given that this column is our response variable (the categorical class we're predicting), hence why this is an important step. An additional touch was that we made all the coloumns containing numbers into integers. 

```r
#pre-process the data by splitting into train/test set

train_test_split<-initial_split(process_hd_data, prop=0.70, strata=Diagnosis) #Using 70% as training set and 30% as testing.
# Assign name to each set
train_set<-training(train_test_split)
test_set<-testing(train_test_split)
```

Here we have pre-processed the cleaned up data by splitting our filtered Hungarian data into the test and training set. We chose to use 70% for our training and 30% for the testing. This is a preparation step for training and testing the classifier that we are going to build.

```r
#calculates the count and percentages of the number of observations for each type of the Diagnosis variable in the process_hd_data data frame.
#additionally finds out the mean of each predictor variable for the Healthy and Unhealthy diagnosis. 

summary<- train_set |>
    group_by(Diagnosis) |>
  summarize(across(c(age, chol, trestbps, thalach, oldpeak), mean)) 
cat("Table 3: Number of Healthy/Unhealthy observations and their percentage")

num_obs <- nrow(process_hd_data)
process_hd_data |>
group_by(Diagnosis) |>
summarize(count = n(),percentage = n() / num_obs * 100)

cat("\n\nTable 4: The average values of the variables")
summary
```

Here we have summarized our data to find the counts and percentages of both healthy and unhealthy diagnoses in our filtered data frame. We adittionaly found the average values for all of the vpredictor ariables in our refined and filtered data, for both diagnoses (healthy and unhealthy). This process was in order to give us a sense of the spread of our data, as well as the spread of our predictor variables, per each diagnoses. 

```r
# create visualization of showing the distribution of Diagnosis based on column that has high-variance in value 
options(repr.plot.width = 15, repr.plot.height = 5)

# Scatterplot for age variables and the resting blood pressure
plot_1 <- ggplot(process_hd_data, aes(x = age, y = trestbps, color = Diagnosis)) +
  geom_point(alpha=3) +
  labs(title = "Graph 1: Scatterplot of Age against Serum Cholestoral levels",
       x = "Age(years)", y = "Resting Blood Pressure (mmhg)", color = "Diagnosis") +
geom_smooth(method = "loess", fill = "blue", alpha = 0.2)+
  theme(text = element_text(size = 12))



# Scatterplot for cholestoral level variables and the maxium heart rate 
plot_2 <- ggplot(process_hd_data, aes(x = chol, y = thalach, color = Diagnosis)) +
  geom_point(alpha=3) +
  labs(title = "Graph 2: Scatterplot of Resting Blood Pressure (mmhg) against Maximum Heart Rate (per min)",
       x = "Serum Cholestoral levels(mg/dl)", y = "Maximum Heart Rate (per min)", color = "Diagnosis") +
  theme(text = element_text(size = 12))


grid.arrange(plot_1,plot_2, nrow = 1)
```

After producing multiple plots with the variables within our data, we found that resting blood pressure, vs. age, and max heart rate vs. serum cholesterol levels gave us the most most visible and meaningful relationships. Within graph 1, we see how as age increases, not only does resting blood pressure increase, but there is a visible higher trend in increase of unhealthy diagnosis for heart disease, meaning the patient does have heart disease (positive relationship). For graph 2 which represents  max heart rate vs. serum cholesterol levels, we can see that the lower the serum cholesterol levels are (which is the amount of total cholesterol in their blood), there is more healthy diagnosis, meaning the patient does not have heart disease; however, as serum cholesterol levels increase, and max heart rate seems to decreases, we can see more unhealthy diagnosis (negative relationhip).

```r
#Prevent randomization of data by setting seed number to a fixed number
set.seed(1234)

#Creating standardized recipe from training set
all_k_recipe <- recipe(Diagnosis ~ trestbps+thalach+age+chol+oldpeak , data = train_set) |>
   step_scale(all_predictors()) |>
   step_center(all_predictors())

# building the basic model specification to auto-tune k value
all_k_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
            set_engine("kknn") |>
            set_mode("classification")

# fold 10 times for cross-validation
all_k_vfold<- vfold_cv(train_set, v = 10, strata = Diagnosis)

kval <- tibble(neighbors = c(1:30))

# Workflow analysis for all k values from 1 to 30
all_k_results <- workflow() |>
      add_recipe(all_k_recipe) |>
      add_model(all_k_spec) |>
      tune_grid(resamples = all_k_vfold, grid = kval)|>
#collecting metrics and rank the k values based on their accuracy
      collect_metrics() |>
      filter(.metric == "accuracy") |>
      select(neighbors, mean) |>
      arrange(desc(mean))
cat("Table 5: Highest model accuracy")
head(all_k_results)
```

Here we produced a standerdized recipe, then built our model specification in order to auto-tune the k value. In order to find the best k, we perfomred cross validation, where we esentially split the training set into a bunch of small sets (10), where 1/10th was used for testing, and 9/10 for training. A model was built with the 9 training sets and evaluated against the 1 testing set. This process was repeated 10 times due to having 10 folds, using a different testing set each time. We then chose to test k values between 1-30 in our workflow, and then we collected the metrics to find which k value gave us the highest accuracy of classification. Note that the highest accuracy was given by the value 21, and 22, as they have the same accuracy so either value would work just fine.

```r
#ploting the K_Value vs Accuracy graph to show the relationship between number of k and its accuracy
k_val_plot <- 
       ggplot(all_k_results, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       theme_minimal()+
       labs(x = "Neighbors", y = "Accuracy Estimate",title="Graph 3: Accuracy vs k_value") +
       scale_x_continuous(breaks = seq(2, 30, by = 1))   
options(repr.plot.width = 9, repr.plot.height = 6)
k_val_plot
```

Producing the accuracy graph bettween the range we chose from 1-30 enables us to visually see which value of k gives us the highest accuracy. From here we can conclude the value is k = 21 or k = 22 as again they both show the same level of accuracy. 

```r
# According to table, choose the best k where the accuracy is at most around 0.788
#verify
best_k <- all_k_results |>
        arrange(desc(mean)) |>
        head(1) |>
        pull(neighbors)
best_k
```

```r
#Use the best k_value which we chose k=21 to build final model
best_k_spec<-nearest_neighbor(weight_func="rectangular", neighbors = best_k)|>
    set_engine("kknn")|>
    set_mode("classification")

best_k_workflow<-workflow()|>
    add_recipe(all_k_recipe)|>
    add_model(best_k_spec)|>
    fit(data = train_set)

cat("Table 6: Model Worflow")
best_k_workflow
    
```

After now knowing which value our best k is (k = 21), we made our final model using it, and created our new workflow with in in addition to training the classifier with the training set. Notice that our recipe stayed the same, as there are no changes to be made to the preprocessing of our data.

```r
#Apply our best_k_value model to the test set 
set.seed(1234)
test_pred <- predict(best_k_workflow, test_set) |>
               bind_cols(test_set)

#Build confusion matrics to show the truth table for Hungarian testing set

conf_mat<- test_pred |>
    conf_mat(truth = Diagnosis, estimate = .pred_class)
cat("Table 7: Confusion Matrix")
conf_mat

#Build metrics for accuracy
result_matrics<- test_pred|>
    metrics(truth = Diagnosis, estimate = .pred_class) |>
    filter(.metric == "accuracy") 

#Produce precision and recall table from out test_set result
recall1<-test_pred |>
    recall(truth = Diagnosis, estimate = .pred_class, event_level="second")


predictions1<-test_pred |>
    precision(truth = Diagnosis, estimate = .pred_class, event_level="second")

cat("\n\nlevels of diagnoses")
test_pred |> pull(Diagnosis) |> levels()

cat("\nTable 8: Recall, precision, and recall of the model")
combined_results <- bind_rows(recall1, predictions1, result_matrics)
combined_results
```

Given our final model, we applied it to our test set. We then built our confusion matrix in order to see how well our classifier performed. To actually calulcate how well the classifier was, we calcualted the accuracy, precision, and recall and placed the data into a chart. We can see that our recall is very low, meaning that we missed identifying a large number of positive diagnoses from the data set (missed a lot of unhealthy diagnoses). However, we have a relatively high precision, meaning out of the diagnosis that our classifier labelled as Unhealthy, 78% of those were actually Unhealthy (meaning people with heart disease).

### Now in order to visualize the performance of our classifier, we produced a visualization of the confusion matrix (our data analysis): 

```r
prediction_matrix_tbl <- tidy(conf_mat)

prediction_matrix_tbl[1, "name"] <- "True Negative" 
prediction_matrix_tbl[2, "name"] <- "False Positive" 
prediction_matrix_tbl[3, "name"] <- "False Negative" 
prediction_matrix_tbl[4, "name"] <- "True Positive" 
prediction_matrix_tbl
```

Given that our confusion matrix isn't really set out in a data table that we usually work with, we first do that in this step so that we can properly complete further analysis and work on it.

```r
matrix_plot_data <- uncount (tibble(Truth = "Healthy(-)", Prediction = "Healthy (-)"), pull(prediction_matrix_tbl[1,2]))|>
    bind_rows(uncount (tibble(Truth = "Healthy(-)", Prediction = "Unhealthy (+)"), pull(prediction_matrix_tbl[2,2]))) |>
    bind_rows(uncount (tibble(Truth = "Unhealthy(+)", Prediction = "Healthy (-)"), pull(prediction_matrix_tbl[3,2]))) |>
    bind_rows(uncount (tibble(Truth = "Unhealthy(+)", Prediction = "Unhealthy (+)"), pull(prediction_matrix_tbl[4,2]))) 

tail(matrix_plot_data)
```

In this step, we have now created a table containing our Truth and Prediction as coloumns, where each row represent a combination of the observations found within the origional confusion matrix. Formatting the data in this way will enable us to use the infromation within it to actually make our graph, as it's no longer just a representation of the data from the confusion matrix, but is actually each combination showed by the matrix itself, represented in each row. 

```r
options(repr.plot.width = 8, repr.plot.length = 8)
prediction_matrix_plot <- matrix_plot_data |>
    ggplot() + 
    geom_mosaic (aes(x = product(Truth, Prediction), fill = Prediction)) +
    labs (title = "Graph 3:Prediction of Diagnosis for heart Disease") + 
    theme(text = element_text(size = 12))


updated_plot <- prediction_matrix_plot +
    geom_text(data = ggplot_build(prediction_matrix_plot)$data[[1]]|>
             mutate(L = c("True Negative", "False Negative", "False Positive", "True Positive")), 
             aes(x = (xmin + xmax)/2, y = (ymin + ymax)/2, label = L))
updated_plot  
```

The mosaic plot above shows the relationship that we earlier discovered within our confusion matrix, however now it visually shows the proportions which make up the matrix. 

First lets look at the negative cases. We can see that for a good portion, we were able to identify the true healthy diagnosis (true negative ~ people who we classified to not have heart disease, which truthfully didn't have heart disease). The false negative portion is smaller, but still noticeably large. This block represents the diagnoses which was classfied as healthy, but was acually unhealthy. That is esentially it is the misdiagnosis for patients who actually had heart disease. 

Now lets take a look at the positive cases. The true positive portion is quite larger in comparisson to the false positive. Meaning that we were able to properly classify a decent number of unhealthy cases as unhealthy (meaning properly diagnose people with heart disease). The false positives, which is the small cube on the bottom right of the graph is representative of the patients who were diagnosed as unhealthy, but were actually healthy (misdiagnosis of having heart disease, when they don't). 

### Additional testing of the classifier:

```r
#Testing model on cleveland data set

cle_data <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data",
                      col_names = c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"))

cat("Table 9: Cleveland Data Unfiltered")
head(cle_data)
```

```r
#reprocess our data so that num column has class value of 0 and 1
new<-cle_data|>
mutate(num = ifelse(num != 0, 1, num))
cat("Table 10: Cleveland data with Added Diagnosis Coloumn")
head(new)

#Choosing the predicting variables that we are going to use to predict the class type(avoid using slop,ca,thl because they have missing values)
process_cl_data<-new|>
#choosing the predicting variables and class variable
select(c(age,chol,trestbps,thalach,oldpeak,num))|> 
#filter out any missing values
filter(age!="?",chol!="?",trestbps!="?",thalach!="?",oldpeak!="?",num!="?")|>
#change the type of the data to dbl because the predicting variables we predicting should be double
mutate(across(c(age, chol, trestbps, thalach, oldpeak, num), as.integer))|>
#creating new column from num variable for better understanding about our classifier
mutate(Diagnosis= case_when(num == 0 ~ "Healthy",num == 1 ~ "Unhealthy"))|>
#change Diagnosis to factor because the class variable we predicting should be categorical
mutate(Diagnosis=as.factor(Diagnosis))
process_cl_data$oldpeak <- round(process_cl_data$oldpeak)
print(20)

cat("Table 11: Filtered Cleveland Data")
head(process_cl_data)
```

In order to test our classifier further, we decided to do extra work to test it on the cleveland data. Here we have preprocessed it, getting rid of any uknowns and adding our new Diagnosis column. Note that within this particular data frame, values 1,2,3, and 4 corresponded with a positive (unhealthy) diganoses, hence we began with reprocessing our data so we only had 0 (healthy), and 1 (unhealthy). 

```r
#Apply hungarian data on cleveland data set

cle_pred <- predict(best_k_workflow, process_cl_data)|>
                      bind_cols(process_cl_data)

cat("Table 12: Predictions for Cleveland Data")
head(cle_pred)

# Convert Prognosis column to factor
cle_pred$Diagnosis <- as.factor(cle_pred$Diagnosis)


#Build metrics for accuracy
cle_matrics<- cle_pred|>
    metrics(truth = Diagnosis, estimate = .pred_class) |>
    filter(.metric == "accuracy") 

#Build confusion matrix to show the truth table for Cleveland dataset
cle_conf_mat<- cle_pred |>
    conf_mat(truth = Diagnosis, estimate = .pred_class)
cat("\n\nTable 13: Confusion Matrix")
cle_conf_mat

#Produce precision and recall table for better understanding
recall2<-cle_pred |>
    recall(truth = Diagnosis, estimate = .pred_class, event_level="second")


predictions2<-cle_pred |>
    precision(truth = Diagnosis, estimate = .pred_class, event_level="second")

cle_combined_results <- bind_rows(recall2, predictions2, cle_matrics)

cat("\n\nTable 14: Precision, Recall, and accuracy for the model")
cle_combined_results
```

What we have done here is made predictions based off of our classifier on the cleveland data, and ultimately made a confusion matrix to then find the values for the recall, precision, and accuracy. We can see that again, our precision is the highest while our recall is the lowest. This means that we missed identifying a big number of positive diagnosis from the data set (missed a lot of actual/true unhealthy diagnoses). However, we have a relatively high precision, meaning out of the diagnosis that our classifier labelled as Unhealthy, 88% of those were actually Unhealthy (meaning people with heart disease).
